```
import pandas as pd
pd.set_option('display.max_rows', None)
metadados = pd.DataFrame.from_records(pd.read_csv("Metadados.csv"))
metadados
```

```
import os
from PIL import Image
import numpy as np
print("Número de classes |",len(np.unique(metadados["classe"])))
print("Número de imagens |",metadados.shape[0])
im = Image.open("/content/drive/Shared drives/EP01 - Visão e Processamento de Imagens/Dataset/1.jpeg")
width, height = im.size
print("Resolução das imagens |",width,"linhas por",height,"colunas")
contador=0
for i in range(1,1081):
    contador=contador+os.stat("/content/drive/Shared drives/EP01 - Visão e Processamento de Imagens/Dataset/{indice_imagem}.jpeg".format(indice_imagem=i)).st_size
print("Tamanho da base em bytes |",contador)
```

```
obj=np.unique(metadados["classe"])
for i in obj:
    classe=metadados[metadados["classe"]==i]
    quant_obj=len(np.unique(classe["Instância"]))
    fundos=np.unique(classe["Fundo"])
    quant_fun=len(np.unique(classe["Fundo"]))
    quant_obs=classe.shape[0]
    s=""
    for j in range(quant_fun):
        if j!=quant_fun-1:
            s=s+fundos[j]+", "
        else:
            s=s+fundos[j]
            
    print(i,"|",quant_obj,"objetos |",quant_fun, "variacoes de fundo:",s
          ,"| 4 variacoes de iluminacao: dia/interior, dia/exterior, noite/interior, noite/exterior |"
          ,quant_obs,"observacoes \n")
pd.options.display.max_colwidth=150
```
